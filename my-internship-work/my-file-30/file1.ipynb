```python
def extract_processor(processor):
    if "Intel Core i3" == processor or "Intel Core i5" == processor or "Intel Core i7" == processor or "Intel Core i9" == processor:
        return processor
    elif "Intel" in processor:
        return "Intel"
    else:
        return "AMD"
```
